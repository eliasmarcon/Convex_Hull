```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
resultsQuickHull = pd.read_csv('../../test_csv_files/Testfile_Quickhull.csv')
resultsQuickHull
```

```python
resultsGiftWrapping = pd.read_csv('../../test_csv_files/Testfile_GiftWrapping.csv')
resultsGiftWrapping
```

```python
# Calculate the expected execution times (O(n * log(n)))
expected_execution_time = resultsQuickHull['Number_of_Points'] * np.log(resultsQuickHull['Number_of_Points'])

# Create a subplot with two plots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Plot the values in a line graph
ax1.plot(resultsQuickHull['Number_of_Points'], resultsQuickHull['Execution Time'], marker='o', label='Actual Execution Time')
ax1.set_xlabel('Number of Points (n)')
ax1.set_ylabel('Execution Time (seconds)')
ax1.set_title('Actual Execution Time vs. Number of Points')
ax1.grid(True)
ax1.legend()

# Plot the O(n * log(n)) example
ax2.plot(resultsQuickHull['Number_of_Points'], expected_execution_time, linestyle='--', label='O(n * log(n))')
ax2.set_xlabel('Number of Points (n)')
ax2.set_ylabel('Execution Time (O(n * log(n)))')
ax2.set_title('Theoretical O(n * log(n)) Execution Time')
ax2.grid(True)
ax2.legend()

# Adjust spacing between plots
plt.tight_layout()

# Show the plots
plt.show()
```

```python
expected_execution_time = resultsGiftWrapping['Number_of_Points'] * resultsGiftWrapping['Number_of_Points']

# Create a subplot with two plots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Plot the values in a line graph
ax1.plot(resultsGiftWrapping['Number_of_Points'], resultsGiftWrapping['Execution Time'], marker='o', label='Actual Execution Time')
ax1.set_xlabel('Number of Points (n)')
ax1.set_ylabel('Execution Time (seconds)')
ax1.set_title('Actual Execution Time vs. Number of Points')
ax1.grid(True)
ax1.legend()

# Plot the O(n * log(n)) example
ax2.plot(resultsGiftWrapping['Number_of_Points'], expected_execution_time, linestyle='--', label='O(n²)')
ax2.set_xlabel('Number of Points (n)')
ax2.set_ylabel('Execution Time O(n²)')
ax2.set_title('Theoretical O(n²) Execution Time')
ax2.grid(True)
ax2.legend()

# Adjust spacing between plots
plt.tight_layout()

# Show the plots
plt.show()
```
